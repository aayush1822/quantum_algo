```python
from qiskit import*
```

```python
circuit=QuantumCircuit(3,3)
```

```python
circuit.draw()
```

```python
circuit.x(0)
circuit.barrier()
```

```python
circuit.draw()
```

```python
circuit.h(1)
circuit.cx(1,2)
circuit.cx(0,1)
circuit.h(0)
```

```python
circuit.draw()
```

```python
circuit.barrier()
circuit.measure(0,0)
circuit.draw()
```

```python
circuit.measure(1,1)
```

```python
circuit.draw()
```

```python
circuit.cx(1,2)
```

```python
circuit.cz(0,2)
#circuit.z(0,1)
circuit.measure(2,2)
```

```python
circuit.draw()
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(circuit,backend=simulator, shots=2048).result()
```

```python
counts=result.get_counts()
from qiskit.tools.visualization import plot_histogram
```

```python
plot_histogram(counts)
```
